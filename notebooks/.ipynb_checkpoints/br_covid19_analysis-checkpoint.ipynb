```python
# Importar bibliotecas PySpark
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, 
    when, 
    isnan, 
    count, 
    to_date, 
    year, 
    month, 
    dayofmonth, 
    weekofyear, 
    sum as spark_sum
)
from pyspark.sql.types import IntegerType, FloatType, StringType

# Bibliotecas para visualização
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import folium
from folium.plugins import HeatMap

# Configurar visualizações no Jupyter
%matplotlib inline
sns.set(style="whitegrid")
```

```python
# Iniciar uma sessão Spark
spark = SparkSession.builder \
    .appName("BR_COVID19_Analysis") \
    .getOrCreate()
```

```python
# Caminho para o arquivo CSV (ajuste conforme o local onde você baixou os dados)
covid_data_path = "caso_full.csv"

# Carregar os dados no DataFrame Spark
covid_df = spark.read.csv(covid_data_path, header=True, inferSchema=True)

# Exibir as primeiras linhas
covid_df.show(5)
```

```python
# Mostrar o esquema do DataFrame
covid_df.printSchema()

# Obter informações gerais sobre os dados
covid_df.describe().show()

# Contagem total de registros
total_cases = covid_df.count()
print(f"Total de registros: {total_cases}")
```

```python
# Renomear colunas para nomes consistentes (minúsculas e sem espaços)
from pyspark.sql.functions import col

covid_df = covid_df.select([col(c).alias(c.strip().lower()) for c in covid_df.columns])

# Verificar os novos nomes das colunas
print(covid_df.columns)

```

```python
# Obter tipos de dados das colunas
print(covid_df.dtypes)
```

```python
# Definir listas de colunas numéricas e não numéricas
numeric_cols = [c for c, dtype in covid_df.dtypes if dtype in ('int', 'double')]
non_numeric_cols = [c for c in covid_df.columns if c not in numeric_cols]
```

```python
from pyspark.sql.functions import count, when, isnan

# Criar expressões para contar valores nulos
null_counts = [count(when(col(c).isNull(), c)).alias(c) for c in non_numeric_cols] + \
             [count(when(col(c).isNull() | isnan(c), c)).alias(c) for c in numeric_cols]

# Selecionar e mostrar os contadores de valores nulos
covid_df.select(null_counts).show()
```

```python
# Remover registros com valores nulos nas colunas críticas
covid_df = covid_df.dropna(subset=["date", "state", "new_confirmed", "new_deaths"])
```

```python
# Mostrar o esquema atualizado
covid_df.printSchema()
```

```python
# Converter colunas para tipos numéricos, se necessário
from pyspark.sql.types import IntegerType, DoubleType

covid_df = covid_df.withColumn("new_confirmed", col("new_confirmed").cast(IntegerType())) \
                   .withColumn("new_deaths", col("new_deaths").cast(IntegerType())) \
                   .withColumn("epidemiological_week", col("epidemiological_week").cast(IntegerType())) \
                   .withColumn("estimated_population", col("estimated_population").cast(IntegerType())) \
                   .withColumn("estimated_population_2019", col("estimated_population_2019").cast(IntegerType())) \
                   .withColumn("is_last", col("is_last").cast(IntegerType())) \
                   .withColumn("is_repeated", col("is_repeated").cast(IntegerType())) \
                   .withColumn("last_available_confirmed", col("last_available_confirmed").cast(IntegerType())) \
                   .withColumn("last_available_confirmed_per_100k_inhabitants", col("last_available_confirmed_per_100k_inhabitants").cast(DoubleType())) \
                   .withColumn("last_available_death_rate", col("last_available_death_rate").cast(DoubleType())) \
                   .withColumn("last_available_deaths", col("last_available_deaths").cast(IntegerType())) \
                   .withColumn("order_for_place", col("order_for_place").cast(IntegerType()))
```

```python
# Reexecutar a contagem de valores nulos
covid_df.select(null_counts).show()
```

```python
from pyspark.sql.functions import sum as spark_sum

# Agrupar por data e somar novos casos e mortes
daily_trends = covid_df.groupBy("date") \
                       .agg(spark_sum("new_confirmed").alias("total_confirmed"),
                            spark_sum("new_deaths").alias("total_deaths")) \
                       .orderBy("date") \
                       .toPandas()

# Plotar a evolução dos casos confirmados
plt.figure(figsize=(14,7))
sns.lineplot(x="date", y="total_confirmed", data=daily_trends, label='Casos Confirmados')
plt.title("Evolução Diária dos Casos Confirmados de COVID-19 no Brasil")
plt.xlabel("Data")
plt.ylabel("Número de Casos Confirmados")
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Plotar a evolução das mortes
plt.figure(figsize=(14,7))
sns.lineplot(x="date", y="total_deaths", data=daily_trends, color='red', label='Mortes')
plt.title("Evolução Diária das Mortes por COVID-19 no Brasil")
plt.xlabel("Data")
plt.ylabel("Número de Mortes")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Exibir as primeiras linhas do DataFrame agregado
daily_trends.head()
```

```python
# Comparação de Casos e Mortes por Estado
state_summary = covid_df.groupBy("state") \
                        .agg(
                            spark_sum("new_confirmed").alias("total_confirmed"),
                            spark_sum("new_deaths").alias("total_deaths")
                        ) \
                        .orderBy(col("total_confirmed").desc()) \
                        .toPandas()

# Calcular taxa de mortalidade por estado
state_summary['mortality_rate'] = (state_summary['total_deaths'] / state_summary['total_confirmed']) * 100

# Exibir os 10 estados com mais casos
top_states = state_summary.head(10)

# Plotar os top 10 estados com mais casos confirmados com a correção do aviso
plt.figure(figsize=(14,7))
sns.barplot(
    x="total_confirmed",
    y="state",
    data=top_states,
    palette="viridis",
    hue="state",      # Adiciona 'hue' atribuído a 'state'
    dodge=False       # Garante que as barras não sejam agrupadas
)
plt.title("Top 10 Estados com Mais Casos Confirmados de COVID-19 no Brasil")
plt.xlabel("Número de Casos Confirmados")
plt.ylabel("Estado")
plt.legend([], [], frameon=False)  # Remove a legenda
plt.show()
```

```python
# Plotar as taxas de mortalidade
plt.figure(figsize=(14,7))
sns.barplot(
    x="mortality_rate",
    y="state",
    data=top_states,
    palette="magma",
    hue="state",
    dodge=False
)
plt.title("Taxa de Mortalidade por Estado (Top 10 Estados com Mais Casos)")
plt.xlabel("Taxa de Mortalidade (%)")
plt.ylabel("Estado")
plt.legend([], [], frameon=False)  # Remove a legenda
plt.show()
```
